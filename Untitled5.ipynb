```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

dataset = spark.read.csv('BostonHousing.csv',inferSchema=True, header =True)
```

```
dataset.printSchema()
```

```
#Input all the features in one vector column
assembler = VectorAssembler(inputCols=['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat'], outputCol = 'Attributes')

output = assembler.transform(dataset)

#Input vs Output
finalized_data = output.select("Attributes","medv")

finalized_data.show()
```

```
#Split training and testing data
train_data,test_data = finalized_data.randomSplit([0.8,0.2])


regressor = LinearRegression(featuresCol = 'Attributes', labelCol = 'medv')

#Learn to fit the model from training set
regressor = regressor.fit(train_data)

#To predict the prices on testing set
pred = regressor.evaluate(test_data)

#Predict the model
pred.predictions.show()
```

```
d = d1.union(d2).union(d3).union(d4).union(d5)
```

```
d.show()
```

```
station2.printSchema()
```

```
d_extended = d.join(station2, on = d['STN---'] == station2['STN_NO']).drop(station2['STN_NO']).drop(station2['COUNTRY_ABBR'])
```

```
d_extended.printSchema()
```

```
from pyspark.sql.functions import when, lit, col

def replace(column, value):
    return when(column != value, column).otherwise(lit(None))
```

```
#|| SLP|   STP|VISIB|WDSP|MXSPD| GUST|  MAX|  MIN| PRCP| SNDP
d_extended = d_extended.withColumn("TEMP", replace(col("TEMP"), 9999.9)).withColumn("DEWP", replace(col("DEWP"), 9999.9))
d_extended = d_extended.withColumn("SLP", replace(col("SLP"), 9999.9)).withColumn("STP", replace(col("STP"), 9999.9))
d_extended = d_extended.withColumn("VISIB", replace(col("VISIB"), 999.9)).withColumn("WDSP", replace(col("WDSP"), 999.9))
d_extended = d_extended.withColumn("MXSPD", replace(col("MXSPD"), 999.9)).withColumn("GUST", replace(col("GUST"), 999.9))
d_extended = d_extended.withColumn("MAX", replace(col("MAX"), 9999.9)).withColumn("MIN", replace(col("MIN"), 9999.9))
d_extended = d_extended.withColumn("PRCP", replace(col("PRCP"), 99.9)).withColumn("SNDP", replace(col("SNDP"), 999.9))
```

```
d_extended.show()
```

```
removeAllDF = d_extended.na.drop()
#columns_with_nas = ["TEMP","DEWP", "SLP", "STP", "VISIB", "WDSP", "MXSPD", "GUST", "MAX", "MIN", "PRCP", "SNDP"]
columns_with_nas = ["MAX", "MIN", "PRCP"]

for x in columns_with_nas:
    meanValue = removeAllDF.agg(avg(x)).first()[0]
    print(x, meanValue)
    d_extended= d_extended.na.fill(meanValue, [x])
```

```
d_extended.where(col("MAX").isNotNull()).show()
```

```
df1=d_extended.withColumn("MEAN", (col("MAX")+col("MIN"))/2)
df1.show()
```

```
#Which country had the hottest average mean temperature over the year?

from pyspark.sql import functions as sf
from pyspark.sql.functions import desc
df1.groupby('COUNTRY_FULL').agg(sf.max('MEAN').alias('MAX_MEAN')).sort(desc("MAX_MEAN")).first()[0]
```

```
#Which country had the second highest average mean wind speed over the year?
d_extended.groupby('COUNTRY_FULL').agg(sf.max('WDSP').alias('MAX_WDSP')).sort(desc("MAX_WDSP")).collect()[1][0]
```
